```python
import datetime
import os
import win32com.client
```

```python
path = os.path.expanduser("C:/Users/rajat.chaudhari/Desktop/1")
today = datetime.date.today()
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
inbox = outlook.GetDefaultFolder(6) 
messages = inbox.Items
```

```python
#message.Subject == "some subject"
a=0
for message in messages:
    if  message.Attachments.Count > 0:# and message.Unread:
        # body_content = message.body
        attachments = message.Attachments
        attachment = attachments.Item(1)
        for attachment in message.Attachments:
            
            if str(attachment).split('.')[-1].lower()=='pdf':
                attachment.SaveAsFile(os.path.join(path, str(a)+str(attachment)))
                message.Unread = False
                break
        a+=1
```
